Assignment Operator

```python
x = 5
y = 3


x = 5 # Re-assigning x to 5 for demonstration
print("x after direct assignment (=): x =", x)

# Add and assign +=
x += y # Equivalent to x = x + y
print("x += y: x =", x)

# Subtract and assign -=
x -= y # Equivalent to x = x - y
print("x -= y: x =", x)

# Multiply and assign *=
x *= y # Equivalent to x = x * y
print("x *= y: x =", x)

# Divide and assign /=
x /= y # Equivalent to x = x / y
print("x /= y: x =", x)

# Modulus and assign %=
x %= y # Equivalent to x = x % y
print("x %= y: x =", x)

# Exponent and assign **=
x = 2 # Reset x for a clearer exponentiation example
y = 3
x **= y # Equivalent to x = x ** y
print("x **= y: x =", x)

# Floor divide and assign //=
x = 5 # Reset x
y = 3
x //= y # Equivalent to x = x // y
print("x //= y: x =", x)

# bitwise
x = 5
y = 3

# Bitwise AND and assign &=
x &= y # Equivalent to x = x & y (0101 & 0011 = 0001 = 1)
print("x &= y: x =", x)

x = 5 # Reset x
# Bitwise OR and assign |=
x |= y # Equivalent to x = x | y (0101 | 0011 = 0111 = 7)
print("x |= y: x =", x)

x = 5 # Reset x
# Bitwise XOR and assign ^=
x ^= y # Equivalent to x = x ^ y (0101 ^ 0011 = 0110 = 6)
print("x ^= y: x =", x)

x = 5 # Reset x
# Bitwise right shift and assign >>=
x >>= 1 # Equivalent to x = x >> 1 (0101 >> 1 = 0010 = 2)
print("x >>= 1: x =", x)

x = 5 # Reset x
# Bitwise left shift and assign <<=
x <<= 1 # Equivalent to x = x << 1 (0101 << 1 = 1010 = 10)
print("x <<= 1: x =", x)
```
